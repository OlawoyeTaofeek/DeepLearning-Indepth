## Pydantic Fundamental 

```python
from pydantic import BaseModel, ValidationError, EmailStr
import json
```

```python
class UserInput(BaseModel):
    name: str
    email: EmailStr
    query: str


user_input = UserInput(
                name="John Doe", 
                email="john.doe@example.com", 
                query="What is the capital of France?"
            )

print(user_input)
```

```python
## wrong one
user_input = UserInput(
                name="John User", 
                email="john.doe-@example.com.", 
                query="What is the capital of France?"
            )
```

```python
print(user_input)
```

```python
## wrong one
user_input = UserInput(
                name="John User", 
                email="not-an-@email.com", 
                query="What is the capital of France?"
            )

print(user_input)
```

```python
def validate_user_input(data: dict) -> UserInput:
    try:
        user_input = UserInput(**data)
        print("Valid user input created successfully.")
        print(f"Name: {user_input.model_dump_json(indent=4)}")
        return user_input
    except ValidationError as e:
        print("Validation error:", e)
        for error in e.errors():
            print(f"Error '{error['loc'][0]}': {error['msg']}")
        return None
```
